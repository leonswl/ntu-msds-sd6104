```python
import pandas as pd
from data_profiler import DataProfiler
import matplotlib.pyplot as plt   # only needed for the plots

# ------------------------------------------------------------------
# 1. Load DataFrame
# ------------------------------------------------------------------
df = pd.read_csv("Food_Inspections_20250216.csv")     

# ------------------------------------------------------------------
# 2. Create a profiler instance
# ------------------------------------------------------------------
profiler = DataProfiler(df)
```

```python
# Overall Single-Column Profiling in dataframes:
profile_df = profiler.profile_dataframe()   # returns a tidy DataFrame
profile_df                # inspect results
```

```python
# Single Column profiling details:
city_profile = profiler.profile_column("City")   # any column name
city_profile
```

```python
# Plotting first digit distribution:
plt.figure(figsize=(5,3))
profiler.plot_first_digit("Inspection ID")  # must be numeric‑like
plt.show()
```

```python
# --- Text columns visualise ------------------------------------
cols_to_plot = ['City', 'Facility Type', 'State', 'Results']

for col in cols_to_plot:
    # FULL RANGE ----------------------------------------------------
    plt.figure(figsize=(10, 4))
    # use the exact number of unique values as top_n → shows everything
    profiler.plot_text_frequency(
        col=col,
        top_n=df[col].nunique(dropna=False),  # show all categories
        show_pct=False                        # or True for %
    )
    plt.title(f"{col} – full range")
    plt.show()

    # TOP‑20 (+ Others) --------------------------------------------
    plt.figure(figsize=(10, 4))
    profiler.plot_text_frequency(
        col=col,
        top_n=20,       # collapse the rest into 'Others'
        show_pct=False  # flip to True if you prefer percentages
    )
    plt.title(f"{col} – top 20 categories")
    plt.show()
```
